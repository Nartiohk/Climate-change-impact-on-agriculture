```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import altair as alt
import seaborn as sns

df = pd.read_csv('climate_change_impact_on_agriculture_2024.csv')
df

```

```python
# Lấy năm
current_year = pd.Timestamp.now().year

# Lọc dữ liệu 5 năm trở lại đây
df_within_5_years = df[df['Year'] >= current_year - 5]

# Reset index
df_within_5_years.reset_index(drop=True, inplace=True)

df_within_5_years
```

```python
# Kiểm tra và xử lý giá trị null nếu có
# Kiểm tra giá trị null trong DataFrame
print(df_within_5_years.isnull().sum())

# Xử lý giá trị null (nếu có) bằng cách điền giá trị trung bình cho các cột số
for col in df_within_5_years.select_dtypes(include=['number']):
    df_within_5_years[col].fillna(df_within_5_years[col].mean(), inplace=True)

# Xử lý giá trị null (nếu có) bằng cách điền giá trị xuất hiện nhiều nhất cho các cột phân loại
for col in df_within_5_years.select_dtypes(include=['category', 'object']):
    df_within_5_years[col].fillna(df_within_5_years[col].mode()[0], inplace=True)

# Kiểm tra lại giá trị null sau khi xử lý
print(df_within_5_years.isnull().sum())
df_within_5_years
```

```python
# Kiểm tra và xử lý các giá trị ngoại lai trong các cột số (nếu có)
numerical_cols = df_within_5_years.select_dtypes(include=np.number).columns


for col in numerical_cols:
    # Calculate IQR
    Q1 = df_within_5_years[col].quantile(0.25)
    Q3 = df_within_5_years[col].quantile(0.75)
    IQR = Q3 - Q1

    
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    
    outliers = df_within_5_years[(df_within_5_years[col] < lower_bound) | (df_within_5_years[col] > upper_bound)]

    
    if not outliers.empty:
        
        median_value = df_within_5_years[col].median()

        
        df_within_5_years.loc[outliers.index, col] = median_value
        print(f"Outliers replaced in '{col}' with the median value ({median_value})")
    else:
        print(f"No outliers detected in '{col}'")

df_within_5_years
```

```python
# Đảm bảo tất cả các tỷ lệ phần trăm nằm trong khoảng 0-100


df_within_5_years['Irrigation_Access_%'] = df_within_5_years['Irrigation_Access_%'].clip(lower=0, upper=100)


df_within_5_years['Soil_Health_Index'] = df_within_5_years['Soil_Health_Index'].clip(lower=0, upper=100)


df_within_5_years['Fertilizer_Use_KG_per_HA'] = df_within_5_years['Fertilizer_Use_KG_per_HA'].clip(lower=0, upper=100)

df_within_5_years
```

```python

# Tạo figure với kích thước lớn
plt.figure(figsize=(15, 10))

# Tính toán correlation matrix cho các yếu tố số học
numeric_columns = ['Average_Temperature_C', 'Total_Precipitation_mm', 'CO2_Emissions_MT',
                  'Crop_Yield_MT_per_HA', 'Extreme_Weather_Events', 'Irrigation_Access_%',
                  'Pesticide_Use_KG_per_HA', 'Fertilizer_Use_KG_per_HA', 'Soil_Health_Index',
                  'Economic_Impact_Million_USD']

correlation_matrix = df_within_5_years[numeric_columns].corr()

# Vẽ heatmap
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Biểu đồ nhiệt thể hiện sự tương quan giữa các yếu tố ảnh hưởng đến tác động kinh tế')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Cường độ màu trong biểu đồ thể hiện cường độ của tương quan giữa
hai yếu tố. Màu đỏ sẫm biểu thị tương quan dương mạnh, trong khi màu
xanh lam sẫm biểu thị tương quan âm mạnh. Các ô vuông màu trắng hoặc
gần trắng biểu thị tương quan tuyến tính thấp hoặc không có.
Các số trong mỗi ô vuông biểu thị chính hệ số tương quan, dao động
từ -1 đến +1.
- +1: Tương quan dương hoàn hảo. Khi một yếu tố tăng, yếu tố kia
tăng theo tỷ lệ.
- -1: Tương quan âm hoàn hảo. Khi một yếu tố tăng, yếu tố kia giảm
theo tỷ lệ.
- 0: Không có tương quan tuyến tính. Những thay đổi trong một yếu
tố không dự đoán được những thay đổi trong yếu tố kia.

Khi nhìn vào biểu đồ, ta có thể thấy được ô vuông tại giao điểm của
"Total_Precipitation_mm" và "Crop_Yield_MT_per_HA" có màu đỏ sẫm
với giá trị gần bằng +0,8, thì ô vuông này biểu thị mối tương quan dương
mạnh giữa tổng lượng mưa và năng suất cây trồng. Nghĩa là, lượng mưa
cao hơn có xu hướng liên quan đến năng suất cây trồng cao hơn. Ngược
lại, ô vuông màu xanh lam đậm sẽ biểu thị mối tương quan âm (ví dụ: sử
dụng nhiều thuốc trừ sâu hơn có mối tương quan âm với chỉ số sức khỏe
đất)

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x='Average_Temperature_C', y='Crop_Yield_MT_per_HA', data=df_within_5_years)
plt.title('Ảnh hưởng của nhiệt độ tới năng suất cây trồng')
plt.xlabel('Nhiệt độ trung bình (°C)')
plt.ylabel('Năng suất cây trồng (MT per HA)')
plt.show()
```

Phân tích mối quan hệ: Ở các vùng nhiệt độ thấp (-5°C đến 10°C) có
năng suất tương đối thấp, mức độ phân tán khá đồng đều và có xu hướng
tăng nhẹ khi nhiệt độ tăng. Còn đối với các vùng có nhiệt độ tối ưu (từ
10°C đến 20°C) thì năng suất ở các vùng này đạt cao nhất, có một số điểm
lến đến gần 5MT/HA, những vùng này có phân bố điểm rộng nhất, cho
thấy sự đa dạng trong năng suất, thường tập trung ở các vùng 3-4 MT/HA.
Cuối cùng là các vùng ở nhiệt độ cao (20°C đến 35°C), chúng ta thấy được
năng suất có xu hướng giảm dần, phân bố điểm vẫn rộng nhưng ít điểm
đạt năng suất cao và vùng (30-35°C) thì năng suất chủ yếu vẫn ở mức 1-3
MT/HA.

Qua các thông tin mà biểu đồ đã thể hiện, chúng ta có thể kết luận
răng nhiệt độ tối ưu cho năng suất cây trồng sẽ nằm trong khoảng 10 –
20 °C. Đây là vùng nhiệt độ cho phép cây trồng phát triển tốt nhất. Về tác
động của nhiệt độ thì nhiệt độ quá thấp (<10°C) hoặc quá cao (>20°C) đều9
làm giảm năng suất và nhiệt độ cao sẽ có tác động tiêu cực mạnh hơn nhiệt
độ thấp. Qua đó cho chúng ta thấy được ý nghĩa của việc cần chọn thời vụ
gieo trồng phù hợp để cây trồng phát triển trong vùng nhiệt độ tối ưu.
Trong điều kiện nhiệt độ không thuận lợi, cần có biện pháp kiểm soát nhiệt
độ phù hợp hoặc chọn các gióng cây trồng chịu nhiệt, chịu lạnh phù hợp

```python
# Create a scatter plot to explore the relationship between CO2 emissions and crop yield
chart = alt.Chart(df_within_5_years).mark_circle().encode(
    x='CO2_Emissions_MT',
    y='Crop_Yield_MT_per_HA',
    color='Crop_Type',  # Color points by crop type for better visualization
    tooltip=['Country', 'Region', 'Year', 'CO2_Emissions_MT', 'Crop_Yield_MT_per_HA', 'Crop_Type'] # Add tooltips for detailed information on hover
).properties(
    title='Relationship between CO2 Emissions and Crop Yield',
    width=800,
    height=600
)

# Display the chart
chart
```

Nhìn vào biểu đồ, ta có thể thấy các vùng phát thải CO2 thấp (0 –
10MT) thì năng suất có sự phân tán rộng từ 1-5 MT/HA, có một số điểm
có năng suất cao (4-5 MT/HA) và mật độ điểm khá đồng đều. Còn ở các
vùng phát thải CO2 ở mức trung bình (10-20MT) thì năng suất không có
gì thay đổi nhiều so với vùng CO2 thấp, mật độ điểm vẫn duy trì đều và ở
cùng phát thải CO2 cao vẫn thế, năng suất vẫn dao động trong mức 1-4
MT/HA. Qua đó thấy được giữa CO2 và năng suất cây trồng không có sự
tương quan nào cụ thể, cho thấy CO2 có thể không phải là một yếu tố
quyết định trực tiếp đến năng suất. Do đó chúng ta cũng ko nên quá quan10
ngại về ảnh hưởng trực tiếp của CO2 đến năng suất mà nên tập trung vào
các yếu tố rõ rệt hơn như nhiệt độ, độ ẩm, dinh dưỡng,..

```python
grouped_data = df_within_5_years.groupby('Crop_Type').agg({
    'Total_Precipitation_mm': 'mean',
    'Crop_Yield_MT_per_HA': 'mean'
}).reset_index()


grouped_data = grouped_data.sort_values('Crop_Type')


fig, ax1 = plt.subplots(figsize=(12, 6))


ax1 = sns.lineplot(x='Crop_Type', y='Total_Precipitation_mm', data=grouped_data, marker='o', color='royalblue', label='Lượng mưa (mm)')
ax1.set_ylabel('Lượng mưa trung bình (mm)', color='royalblue')
ax1.tick_params(axis='y', labelcolor='royalblue')


ax2 = ax1.twinx()
sns.lineplot(x='Crop_Type', y='Crop_Yield_MT_per_HA', data=grouped_data, marker='o', color='red', label='Năng suất cây trồng', ax=ax2)
ax2.set_ylabel('Năng suất cây trồng trung bình trên HA', color='red')
ax2.tick_params(axis='y', labelcolor='red')


plt.title('Sự tương quan giữa trung bình lượng mưa và năng suất trên từng loại cây trồng')
fig.tight_layout()

# Show plot
plt.show()

```

Khi nhìn vào biểu đồ, chúng ta có thể thấy rằng sự tương quan
giữa năng suất và lượng mưa là khác nhau đối với từng loại cây, ví dụ như
Corn có lượng mưa trung bình thấp nhất và năng suất cây trồng cũng thấp,
cho thấy loại cây này có thể cần nhiều nước hơn để phát triển tốt, còn đối
với Sugarcane, loại cây trồng này dù có lượng mưa thấp nhưng năng suất
lại cao hơn nhiều các loại cây trồng khác, cho thấy Sugarcane có thể phát
triển tốt ngay cả trong điều kiện ít mưa. Qua đó chúng ta có thể thấy lượng
mưa cũng giống như CO2, không có tương quan quá nhiều đến năng suất
chung của các loại cây trồng
